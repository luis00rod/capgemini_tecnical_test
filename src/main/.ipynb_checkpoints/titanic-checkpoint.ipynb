Estas transformaciones limpian los datos, preparan las variables para modelos de ML y crean los conjuntos de entrenamiento y prueba necesarios.

# Visualización Exploratoria de los Datos de Titanic

Este código genera varios gráficos para explorar visualmente el dataset de Titanic de Kaggle:

- Gráfico de barras para la distribución de la variable objetivo `Survived` (sobrevivió al hundimiento del Titanic).

- Gráficos de barras estratificados por `Survived` para varias características de los pasajeros (`Pclass`, `Sex`, etc).

- Histograma estratificado por `Survived` para la edad de los pasajeros (`Age`).

Estos gráficos permiten tener una comprensión inicial de la distribución de las variables en el dataset, identificar valores atípicos y relaciones entre características de los pasajeros y la variable objetivo de supervivencia.

```python
def plotter_exploration():
    sns.set_style("whitegrid")
    plt.figure(figsize=(6,4))
    sns.countplot(x='Survived', data=raw_data)
    plt.title('Distribución de Sobrevivientes')
    plt.show()
    features = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']

    for feature in features:
        plt.figure(figsize=(6,4))
        sns.countplot(x=feature, hue='Survived', data=raw_data)
        plt.title(f'Sobrevivientes en función de {feature}')
        plt.show()

    plt.figure(figsize=(6,4))
    sns.histplot(raw_data[raw_data['Survived'] == 0]['Age'], bins=30, color='red', label='No Sobrevivió')
    sns.histplot(raw_data[raw_data['Survived'] == 1]['Age'], bins=30, color='green', label='Sobrevivió')
    plt.title('Distribución de Edad para Sobrevivientes vs No Sobrevivientes')
    plt.legend()
    plt.xticks(np.arange(0, 80+10, 10))
    plt.show()
```

```python
plotter_exploration()
```

# Gráfica de densidad kernel (KDE)

Este código define una función `plot_kde` para graficar la densidad kernel de una variable continua estratificando por la variable categórica objetivo.

- Se filtran los datos de cada clase objetivo.
- Se grafica la densidad KDE de cada clase en el mismo gráfico con diferentes colores. 
- Se definen título, límites de eje x y otras opciones de gráfico.

Luego se utiliza la función para graficar la distribución de la variable `Age` estratificada por la variable objetivo `target`.

Esto permite visualizar la distribución de variables continuas comparando entre clases. 

```python
# Function to plot KDE
def plot_kde(_clean_data, fig, nrows, ncols, mypos, title, feature, xmin=0, xmax=None):
    train_survive = _clean_data[_clean_data.target == 1]
    train_dead = _clean_data[_clean_data.target == 0]
    ax = fig.add_subplot(nrows, ncols, mypos)
    train_survive[feature].plot.kde(ax=ax, color='g')
    train_dead[feature].plot.kde(ax=ax, color='r')
    ax.set_title(title)
    ax.set_xlim(xmin, xmax)

# Plotting
fig = plt.figure(figsize=(12, 10))

plot_kde(clean_data, fig, 2, 2, 1, 'Age', 'Age', xmax=clean_data.Age.max())
plt.tight_layout()
plt.show()
```
